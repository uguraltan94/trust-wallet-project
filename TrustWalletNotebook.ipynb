```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("Senior Data Analyst Task Transactions.csv")
df
```

```python
df.info()
df.describe()
print("Missing values:\n", df.isnull().sum())
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'], errors='coerce')
df['Transaction_Pure_Date'] = df['Transaction_Date'].dt.date
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Day'] = df['Transaction_Date'].dt.day
df['Month_Name'] = df['Transaction_Date'].dt.strftime('%B')
```

```python
df
```

```python
wallet_groups = {
    'Trezor': 'Hardware Wallet',
    'Ledger': 'Hardware Wallet',
    'Cold Wallet': 'Cold Wallet',
    'Hot Wallet': 'Hot Wallet',
    'Exchange Wallet': 'Exchange Wallet'
}
df['Wallet_Category'] = df['Wallet_Type'].map(wallet_groups).fillna('Other')
df
```

```python
wallet_high_level = {
    'Trezor': 'Cold Wallet',
    'Ledger': 'Cold Wallet',
    'Cold Wallet': 'Cold Wallet',
    'Hot Wallet': 'Hot Wallet',
    'Exchange Wallet': 'Hot Wallet'
}
df['Wallet_High_Level_Category'] = df['Wallet_Type'].map(wallet_high_level).fillna('Other')
```

```python
conn = sqlite3.connect(":memory:")
df.to_sql("crypto_transactions_clean", conn, index=False, if_exists="replace")

# SQL query
query = """
SELECT count(distinct user_id) as users, 
count(transaction_id) as total_transactions, 
round(count(case when status = 'Completed' then transaction_id end),2)/round(count(transaction_id),2) 
as total_transactions_completed_percent, 
sum(amount) as volume,
Transaction_Type
FROM crypto_transactions_clean;
"""

# Execute query
result = pd.read_sql(query, conn)
print(result)
```

```python
#Checking Duplicate Transactions which could have gone from Pending to Completed as an example
conn = sqlite3.connect(":memory:")
df.to_sql("crypto_transactions_clean", conn, index=False, if_exists="replace")

# SQL query
query = """
SELECT * FROM (
SELECT *, row_number() OVER (partition by user_id, transaction_id ORDER by transaction_pure_date desc) as rn
FROM crypto_transactions_clean)
WHERE RN = 1;
"""

# Execute query
result = pd.read_sql(query, conn)
print(result)
```
